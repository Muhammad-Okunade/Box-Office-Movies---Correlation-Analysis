```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
# data types

movies.dtypes
```

```python
movies.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Visualising data
sns.regplot(x="gross", y="budget", data=movies, line_kws={'color':'red'})
```

```python
corr_matrix =  movies.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot = True)
plt.title ('Correlation Matrix')
plt.xlabel('Movie Features - Numeric')
plt.ylabel('Movie Features - Numeric')
plt.show()
```
